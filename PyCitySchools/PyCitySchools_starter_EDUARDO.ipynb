```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_df_path = "Resources/schools_complete.csv"
student_df_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_df_path)
student_data = pd.read_csv(student_df_path)

# Combine the data into a single dataset.  
fullschool_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create the variables for the summary table data frame

#Total schools
total_schools = len(fullschool_df["school_name"].unique())

#Total students
total_students = len(fullschool_df["Student ID"].unique())

#Total Budget
total_budget = school_data["budget"].sum()

#avergae math score
avg_math_scr = fullschool_df["math_score"].mean()

#avergae reading score
avg_read_scr = fullschool_df["reading_score"].mean()

#Passing percentages

passed_math = fullschool_df.loc[fullschool_df["math_score"] >= 70]
pct_math_pass = (len(passed_math["Student ID"].unique())/total_students)*100

passed_read = fullschool_df.loc[fullschool_df["reading_score"] >= 70]
pct_read_pass = (len(passed_read["Student ID"].unique())/total_students)*100

passed_two = pd.merge(passed_math,passed_read,how="inner",on="Student ID",suffixes=("","2"))
passed_two_count = len(passed_two["Student ID"].unique())
pct_two_pass = (passed_two_count/total_students)*100


#printing out all the values in a summary data frame

district_summary = pd.DataFrame({"Total Schools": [total_schools],
                                 "Total Students" : [total_students],
                                 "Total Budget" : [total_budget],
                                 "Average Math Score" : [avg_math_scr],
                                 "Average Readning Score" : [avg_read_scr],
                                 "% Passing Math" : [pct_math_pass],
                                 "% Passing Reading" : [pct_read_pass],
                                 "% Overall Passing" : [pct_two_pass],
                                 })
district_summary["Total Budget"] = district_summary["Total Budget"].map('${:,.2f}'.format)
district_summary["Total Students"] = district_summary["Total Students"].map('{:,d}'.format)

district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
grp_pass_math = passed_math.groupby(["school_name"])
grp_pass_math_1 = grp_pass_math["Student ID"].count()
```

```python
grp_pass_read = passed_read.groupby(["school_name"])
grp_pass_read_1 = grp_pass_read["Student ID"].count()
```

```python
grp_pass_two = passed_two.groupby(["school_name"])
grp_pass_two_1 = grp_pass_two["Student ID"].count()
```

```python
grp_full_data = fullschool_df.groupby(["school_name"])

grp_avg_math = grp_full_data["math_score"].mean()
grp_avg_read = grp_full_data["reading_score"].mean()

school_df = pd.DataFrame({"Average Math Score":grp_avg_math,
                          "Average Reading Score":grp_avg_read,
                          "Passed Math":grp_pass_math_1,
                          "Passed Reading":grp_pass_read_1,
                          "Overall Passed":grp_pass_two_1,
                         })
```

```python
school_2_df = pd.merge(school_data,school_df,on = "school_name")
```

```python
school_2_df["Per Student Budget"] = school_2_df["budget"]/school_2_df["size"]
school_2_df["% Passing Math"] = school_2_df["Passed Math"]/school_2_df["size"]*100
school_2_df["% Passing Reading"] = school_2_df["Passed Reading"]/school_2_df["size"]*100
school_2_df["% Overall Passing"] = school_2_df["Overall Passed"]/school_2_df["size"]*100
```

```python
school_2_df = school_2_df.rename(columns={"school_name" : "School Name",
                                          "type" : "School Type",
                                          "size" : "Total Students",
                                          "budget" : "Total School Budget",
                                          })

school_2_df_clean = school_2_df [["School Name",
                                 "School Type",
                                 "Total Students",
                                 'Total School Budget',
                                 'Per Student Budget',
                                 'Average Math Score',
                                 'Average Reading Score',
                                 '% Passing Math',
                                 '% Passing Reading',
                                 '% Overall Passing']]

school_2_df_clean.loc[:, "Total School Budget"] = school_2_df_clean["Total School Budget"].astype("float")
school_2_df_clean.loc[:, "Per Student Budget"] = school_2_df_clean["Per Student Budget"].astype("float")


school_3_df_clean = school_2_df_clean.set_index("School Name")
school_3_df_clean.index.name = None
school_3_df_clean.loc[:, "Total School Budget"] = school_3_df_clean["Total School Budget"].map('${:,.2f}'.format)
school_3_df_clean.loc[:, "Per Student Budget"] = school_3_df_clean["Per Student Budget"].map('${:,.2f}'.format)


school_3_df_clean
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
n9th_grade = fullschool_df.loc[fullschool_df["grade"]=="9th"]
n10th_grade = fullschool_df.loc[fullschool_df["grade"]=="10th"]
n11th_grade = fullschool_df.loc[fullschool_df["grade"]=="11th"]
n12th_grade = fullschool_df.loc[fullschool_df["grade"]=="12th"]

grp9th = n9th_grade.groupby("school_name")
grp10th = n10th_grade.groupby("school_name")
grp11th = n11th_grade.groupby("school_name")
grp12th = n12th_grade.groupby("school_name")

math9_scores = grp9th["math_score"].mean()
math10_scores = grp10th["math_score"].mean()
math11_scores = grp11th["math_score"].mean()
math12_scores = grp12th["math_score"].mean()

mathscores = pd.DataFrame({"9th" : math9_scores,
                          "10th" : math10_scores,
                          "11th" : math11_scores,
                          "12th" : math12_scores
                          })

mathscores.loc[:, "9th"] = mathscores["9th"].map('{:,.2f}'.format)
mathscores.loc[:, "10th"] = mathscores["10th"].map('{:,.2f}'.format)
mathscores.loc[:, "11th"] = mathscores["11th"].map('{:,.2f}'.format)
mathscores.loc[:, "12th"] = mathscores["12th"].map('{:,.2f}'.format)

mathscores.index.name = None
mathscores


```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins
bins = [0,584.99, 629.99, 644.99, 680]

budget_cat = ">$585", "$585-$630", "$630-$645", "$645-$680"

school_2_df_clean["Spending Ranges (Per Student)"] = pd.cut(school_2_df_clean["Per Student Budget"], bins, labels=budget_cat, include_lowest = True)
budget_spend_summ = school_2_df_clean.groupby("Spending Ranges (Per Student)").mean()

budget_spend_summ = budget_spend_summ[["Average Math Score",
                                      "Average Reading Score",
                                      '% Passing Math',
                                      '% Passing Reading',
                                      '% Overall Passing']]

budget_spend_summ.loc[:, "Average Math Score"] = budget_spend_summ["Average Math Score"].map('{:,.2f}'.format)
budget_spend_summ.loc[:, "Average Reading Score"] = budget_spend_summ["Average Reading Score"].map('{:,.2f}'.format)
budget_spend_summ.loc[:, '% Passing Math'] = budget_spend_summ['% Passing Math'].map('{:,.2f}'.format)
budget_spend_summ.loc[:, '% Passing Reading'] = budget_spend_summ['% Passing Reading'].map('{:,.2f}'.format)
budget_spend_summ.loc[:, '% Overall Passing'] = budget_spend_summ['% Overall Passing'].map('{:,.2f}'.format)

budget_spend_summ
```
